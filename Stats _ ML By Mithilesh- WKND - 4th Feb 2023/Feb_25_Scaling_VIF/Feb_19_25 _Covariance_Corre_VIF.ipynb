# Variance _inflation_factor

```python
# here also we get the information about the multicollinearity
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data1.head()
```

```python
# split the dependent and independent variable
X=data1.drop(["target"],axis=1)
Y=data1["target"]
```

```python
X.head()
```

```python
# .values function convert the dataframe to arrays
X.values
```

```python
Y.head()
```

```python
X=sm.add_constant(X)
```

```python
X.head()
```

```python
vif=pd.DataFrame()# blank dataframe
vif["variables"]=X.columns
vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
```
