```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("spam.csv",encoding="latin-1")
```

```python
df=data.copy()
df.head()
```

```python
df=df[["v2","v1"]].rename(columns={"v2":"email","v1":"label"})
```

```python
df.tail()
```

```python
df.info()
```

```python
df.label.value_counts(normalize=True)
```

```python
# rename the target variable
df["label"]=df["label"].map({"ham":0,"spam":1})
df["label"].value_counts()
```

```python
# dependent and independent variable
```

```python
X=df["email"]
Y=df["label"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
X_train.tail()
```

```python
# creation of Bag of words
```

```python
# we can use the function CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
new_x=vectorizer.fit(X_train)
```

```python
new_x
```

```python
x_train_vectorized=vectorizer.transform(X_train)
```

```python
X_test_vectorized=vectorizer.transform(X_test)
```

```python
x_train_vectorized.dtype
```

```python
x_train_vectorized.shape
```

```python
x_train_vectorized
```

```python
# here we have "x_train_vectorized" as out X_train in vectorised format, Y_train is our dependent variable
```

```python
# lets import the algorithm
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
mnb=MultinomialNB(alpha=0.01)
mnb.fit(x_train_vectorized,Y_train)
```

```python
prediction=mnb.predict(X_test_vectorized)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test,prediction)
```

```python
print(classification_report(Y_test,prediction))
```

```python
confusion_matrix(Y_test,prediction)
```

```python
# creatye a function and ckeck the authenticity of model
```

```python
def enter_text(message):
    if mnb.predict(vectorizer.transform([message]))==1:
        print("This is spam")
    else:
        print("its a normal message")
```
