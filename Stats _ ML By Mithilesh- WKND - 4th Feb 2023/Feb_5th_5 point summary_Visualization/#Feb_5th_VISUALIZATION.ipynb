# count plot

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Gender",data=empdf)
plt.title("Distribution of gender")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="BusinessTravel",data=empdf)
plt.title("Distribution of business travel")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Department",data=empdf)
plt.title("Distribution of Department")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="JobRole",data=empdf,hue="Gender")
plt.title("Distribution of JobType")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Department",data=empdf,hue="Attrition")
plt.title("Distribution of Department")
```

```python
# factorplot
sns.factorplot(x="MaritalStatus",kind="count",data=empdf,hue="Attrition")
```

# BARPLOT


```python
plt.figure(figsize=(7,4))
sns.barplot(x="Department",y="MonthlyRate",data=empdf,hue="Attrition")
plt.title("Distribution of Department")
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x="Department",y="Age",data=empdf,hue="Gender")
plt.title("Distribution of Department")
```

```python
plt.figure(figsize=(7,4))
ax=sns.barplot(x="JobSatisfaction",y="JobRole",data=empdf,hue="Gender")
ax.set(ylabel="JobRole")
plt.xlabel("JobSatisfaction",fontsize=20)
plt.title("Distribution of JobSatisfaction")
```

# histogram

```python
plt.figure(figsize=(7,4))
ax=sns.histplot(x="JobSatisfaction",y="JobRole",data=empdf,hue="Gender")
ax.set(ylabel="JobRole")
plt.xlabel("JobSatisfaction",fontsize=20)
plt.title("Distribution of JobSatisfaction")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="JobRole",data=empdf,hue="Gender")
plt.title("Distribution of JobSatisfaction")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="MonthlyIncome",data=empdf,bins=10,kde=True)
plt.title("Distribution of Monthly")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="Age",data=empdf,bins=10,kde=True)
plt.title("Distribution of AGE")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="Age",data=empdf,bins=10,kde=True,hue="Department")
plt.title("Distribution of AGE")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="MonthlyIncome",data=empdf,bins=10,kde=True,hue="Department")
plt.title("Distribution of AGE")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="MonthlyIncome",data=empdf,bins=10,kde=True,hue="Department",stat='probability' )
plt.title("Distribution of AGE")
```

```python
plt.figure(figsize=(15,8))
ax=sns.histplot(x="Age",data=empdf,bins=10,kde=True,hue="Department",stat="density")
plt.title("Distribution of AGE")
```

# pieplot

```python
empdf["Attrition"].value_counts()
```

```python
plt.figure(figsize=(15,8))
plt.pie(x=empdf["Attrition"].value_counts(),labels=empdf["Attrition"].value_counts().index,autopct="%0.2f%%")
```

```python
plt.figure(figsize=(15,8))
plt.pie(x=empdf["Gender"].value_counts(),labels=empdf["Gender"].value_counts().index,autopct="%0.2f%%")
```

```python
plt.figure(figsize=(15,8))
plt.pie(x=empdf["Department"].value_counts(),labels=empdf["Department"].value_counts().index,autopct="%0.2f%%")
```

```python
plt.figure(figsize=(15,8))
setexplode = [0.1, 0, 0]
plt.pie(x=empdf["Department"].value_counts(),labels=empdf["Department"].value_counts().index,autopct="%0.2f%%",explode=setexplode)
```

# setting explode

```python
plt.figure(figsize=(15,8))
setexplode = [0.05, 0, 0,0.1,0,0.1]
plt.pie(x=empdf["EducationField"].value_counts(),labels=empdf["EducationField"].value_counts().index,autopct="%0.2f%%",explode=setexplode)
```
