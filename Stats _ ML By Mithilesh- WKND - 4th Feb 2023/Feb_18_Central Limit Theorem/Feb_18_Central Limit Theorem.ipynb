```python
sampled_mean200=[]
for i in range(len(cardio2)):
    sampled_mean200.append(cardio2.sample(200).mean())
sns.histplot(sampled_mean200,kde=True)
```

```python
sampled_mean200=[]
for i in range(len(cardio2)):
    sampled_mean200.append(cardio2.sample(n=200).mean())
sns.histplot(sampled_mean200,kde=True)
```

```python
sampled_mean100=[]
for i in range(len(cardio2)):
    sampled_mean100.append(cardio2.sample(n=30).mean())
sns.histplot(sampled_mean100,kde=True)
```

```python
# finding the error 
error=cardio2.mean()-pd.Series(sampled_mean200).mean()
error
```
