```python
def palindrom(a):
    if a == a[::-1]:
        return 'yes'
    else:
        return 'no'

def take_user_input():
    usr_input = input('Do you want to enter new string : ')
    while usr_input != 'no':
        str = input('Enter string : ')
        if palindrom(str) == 'no':
            usr_input = input('Do you want to enter new string : ')
        else:
            return 'yes'
        
 
take_user_input()
```
