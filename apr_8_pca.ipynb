# Applying the PCA concept

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca = pca.fit(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
len(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel(" No of principle component")
plt.ylabel("Explained variance ration or amount of information")
```

```python
pca_with_custom_component = PCA(n_components=15)
```

```python
X_pca_applied = pca_with_custom_component.fit_transform(X_scaled)
```

```python
X_pca_applied
```

```python
len(X_pca_applied)
```

```python
X_pca_applied_df = pd.DataFrame(X_pca_applied)
```

```python
X_pca_applied_df.head()
```

```python
X_pca_applied_df.shape
```

```python
# After applying the PCA re-evaluate the model and see does PCA improving the performance of the model
```

```python
X_train_pca, X_test_pca, Y_train_pca, Y_test_pca = train_test_split(X_pca_applied_df, Y, test_size=0.2, random_state=101, stratify=Y)
```

```python
rf_pca = RandomForestClassifier()
rf_pca.fit(X_train_pca, Y_train_pca)
```

```python
Y_predict_rf_pca = rf_pca.predict(X_test_pca)
```

```python
accuracy_score(Y_test_pca, Y_predict_rf_pca)
```

```python
# We can see the performance of the model has been reduced without PCA (with n_components = 15) accuracy was around 94 
# and now its reduced to 93.

# Lets try with PCA (n_components = 20)
```

```python
pca_with_custom_component_20 = PCA(n_components=20)
```

```python
X_pca_applied_20 = pca_with_custom_component_20.fit_transform(X_scaled)
```

```python
X_pca_applied_20_df = pd.DataFrame(X_pca_applied_20)
```

```python
X_train_pca_20, X_test_pca_20, Y_train_pca_20, Y_test_pca_20 = train_test_split(X_pca_applied_20_df, Y, test_size=0.2, random_state=100, stratify=Y)
```

# Mathematics Behind PCA

```python
# X_scaled ---- Original Scaled DATA
```

```python
# Step 1 : Calculate the covariace matrix
```

```python
covariance_matrix = X_scaled.cov()
```

```python
covariance_matrix
```

```python
covariance_matrix.shape
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(covariance_matrix, annot=True, annot_kws={"size": 7})
```

```python
eigen_value1, eigen_vector1 = np.linalg.eig(covariance_matrix)
```

```python
eigen_value1
```

```python
len(eigen_value1)
```

```python
eigen_vector1
```

```python
len(eigen_vector1)
```

```python
eigen_value1_df = pd.DataFrame(eigen_value1)
```

```python
eigen_value1_df.head()
```

```python
eigen_vector1_df = pd.DataFrame(eigen_vector1)
```

```python
eigen_vector1_df.head(100)
```

```python
# Now we will sort the eigen values and eigen vectors
```

```python
sorted_key = np.argsort(eigen_value1)[::-1][:2]
```

```python
sorted_key
```

```python
eigen_values, eigen_vector = eigen_value1[sorted_key], eigen_vector1[:,sorted_key]
```

```python
eigen_values
```

```python
eigen_vector
```

```python
# Dot product of original data and eigen vectors are the principal component.
# This is the projection step of the original points on to the principal component.
```

```python
principal_components = np.dot(X_scaled,eigen_vector1)
```

```python
principal_components
```

```python
principal_components.shape
```
