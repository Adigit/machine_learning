# Handling Missing data
- fillna()
- dropna()

```python
df = pd.read_csv('datasets/weather_data_nan.csv')

df
```

# fillna()

```python
df.fillna(0)
```

```python
df.fillna({
    'temperature': df['temperature'].median(),
    'windspeed': df['windspeed'].mean(),
    'event': df['event'].mode()[0]
})
```

```python
df.fillna({
    'temperature': df['temperature'].median(),
    'windspeed': df['windspeed'].mean(),
    'event': df['event'].mode()[0]
}, inplace = True)
df
```
