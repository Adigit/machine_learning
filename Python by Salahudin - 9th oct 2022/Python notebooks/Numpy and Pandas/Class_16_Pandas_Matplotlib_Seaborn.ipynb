# Handling missing values

```python
df = pd.read_csv('../../datasets/weather_data_nan.csv')

df
```

```python
# temperature --> median
# windspeed ---> mean
# event --> mode

df.fillna({
    'temperature': df['temperature'].median(),
    'windspeed': df['windspeed'].mean(),
    'event': df['event'].mode()[0]
} , inplace = True)
```

# Histogram

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,83,129]


# 80 -100 ---> normal
# 100- 125 ----> prediabetic
# 125 - above ---> diabetic

plt.hist(blood_sugar)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,83,129]


# 80 -100 ---> normal
# 100- 125 ----> prediabetic
# 125 - above ---> diabetic

plt.hist(blood_sugar , bins = [80,100,125,150])
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
# which gender are more prone to have diabetics ? 
# Which gender usually lies in the prediabetic zone? 
```
