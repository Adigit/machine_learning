# Iterating over a data frame

```python
1. itertuple
2. iterrows
3. iteritems
```

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
df
```

```python
# 1. iter tuple

for i in df.itertuples():
    print(i.Headquarters.lower() , end = '\n\n')
```

```python
# 1. iter rows

for i,j in df.iterrows():
    print(i)
    print(j['Headquarters'].lower() , end = '\n\n')
```

```python
df
```

```python
# 1. iter tuple

for i in df.itertuples():
    if i.Organizations_name == 'ISRO':
        i.Country_name = 'Bharat'
```

```python
# 2. Iter rows 

for i,j in df.iterrows():
    if j['Organizations_name']== 'ISRO':
        j['Country_name'] = 'Bharat'
```

```python
df
```

```python
# 1. iter tuple

for i in df.itertuples():
    if i.Organizations_name == 'ISRO':
        print(i.Country_name)
```

```python
# 2. Iter rows 

for i,j in df.iterrows():
    if j['Organizations_name']== 'ISRO':
        print(j['Country_name'])
```

```python
df
```

```python
for i in df.index:
    print(df.loc[i , 'Country_name'])
```

```python
# 3. Iter items

for i,j in df.iteritems():
    print(j , end = '\n\n')
```

```python
for i in df.columns:
    print(i)
```
