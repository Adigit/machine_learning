```python
# enter the amount
# if the amount is less than 50000 then tell that the amount deposited

{
# if the amount is btw 50000 -- 70000 I want you to raise an error_1
# If the error_1 is raise then ask for pan 

# If the len of pan number is 10 ---> amount deposited
# else raise an error and stop every
}

# If the amont is greater than the 70000 raise and error Over_limit
# if the over limit error is raised ,
# Ask if they want to retry from begining
# if Y then do same thing prviously done


# If N then stop


# Log all the steps as well as errors 
```

```python
class HighAmountError(Exception):
    def __init__(self, msg):
        self.msg = msg
        
        
class OverLimitError(Exception):
    def __init__(self, msg):
        self.msg = msg
        
        
class PANError(Exception):
    def __init__(self, msg):
        self.msg = msg
        
        
class StopError(Exception):
    def __init__(self, msg):
        self.msg = msg
```

```python

f = open('file_handling/bank_records.txt' , 'w')

def Bank_Transfer():
    amount = eval(input('Please enter the amount you want to transfer: '))
    f.write(str(amount))
    try:
        if amount >= 0 and amount <= 50000:
            print('Amount transferred')
            f.write(str('\nAmount transferred'))
        elif amount >= 50000 and amount <= 70000:
            raise HighAmountError ('Please give your PAN number')
        elif amount >= 70000:
            raise OverLimitError ('The amount entered is more than the maximum limit of transaction!')
    except HighAmountError as e:
        try:
            f.write(f'\n{str(e)}')
            PAN_number = input('Please give your PAN number: ')
            if len(PAN_number) == 10:
                print('Amount transferred')
            else:
                raise PANError ('Wrong PAN number!')
        except PANError as e:
            f.write(f'\n{str(e)}')
            raise StopError('Please try after sometime')
                
    except OverLimitError as e:
        f.write(f'\n{str(e)}')
        retry = input('Do you want to try with a less amount? Y/N: ')
        f.write(str(retry))
        if retry == 'Y':
            f.write(str('\nGoing for a retry'))
            Bank_Transfer()
        elif retry == 'N':
            print('Have a good day!')
    
    finally:
        f.close()
    
Bank_Transfer()

```

```python
class error_1(Exception):
    def __init__(s,msg):
        s.msg = msg
        
class error_2(Exception):
    def __init__(s,msg):
        s.msg = msg
        
class error_3(Exception):
    def __init__(s,msg):
        s.msg = msg        
        
def enter_amt():
    return input('Enter an amount to deposit. ')

def enter_pan():
    return input('Enter PAN number. ')

def logs(m):
    with open('bank_logs.txt','w') as f:
        f.write(str(m))

def start():

        amt = enter_amt()
        if eval(amt) < 50000:
            logs(f'Amount Deposit --> {amt}\nThanks for using COD')
            return f'Amount Deposit --> {amt}\nThanks for using COD'
        if eval(amt) >= 50000 and eval(amt) <= 70000:
            return f'error_1-{amt}'
        if eval(amt) > 70000:
            return f'error_over_limit-{amt}'



while 1:
    try:
        c = input('Enter "y" to start COD or "n" to exit. ')
        
        if c.lower() == 'y':
            global a 
            a = start()
            if a.split('-')[0] == 'error_1':
                raise error_1('PAN is required for Amount greater than or equal to 50000 ')
            elif a.split('-')[0] == 'error_over_limit':
                raise error_2(f'Amount - {a.split("-")[1]} Overlimit\nUnable to deposit')
            else:
                print(a)
                break
        elif c.lower() == 'n':
            raise error_2('Thanks for using COD')
        else:
            print('Not a valid option. Please enter correct option')

    except error_2 as e:
        logs(e)
        print(e)
        break
    except error_1 as e1:
        logs(e1)
        print(e1)
        pan_details = enter_pan()
        try:
            if len(pan_details) == 10:
                logs(f'Amount Deposit --> {a.split("-")[1]}\nThanks for using COD')
                print(f'Amount Deposit --> {a.split("-")[1]}\nThanks for using COD')
                break
            else:
                raise error_3(f'Invalid PAN number {pan_details}\nDo you want to continue? ')
                break

        except error_3 as e3:
            logs(e3)
            print(e3)
```

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```
