```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(df['open'])
```

```python
def adf_check(timeseries):
    result = adfuller(timeseries)
    print("************Augmented Dickey Fuller Test************")
    labels = ['ADF Statistic Test', 'p-value', '#Lags', 'No of Observation']
    
    for a, b in zip(result, labels):
        print(b + " : " + str(a))
        
    if result[1] <=0.05:
        print("Strong evidence against null hypothesis and my timeseries is Stationary")
    else:
        print("Weak evidence against null hypothesis and my timeseries is Non-Stationary")  
```

```python
adf_check(df['open'])
```

```python
# Facebook Prophet model : it's mandatory to convert your variable as follow -
# Date - 'ds', open - 'y'
```

```python
df_prophet = df.rename(columns={'date':'ds', 'open': 'y'})
```

```python
df_prophet.head()
```

```python
from prophet import Prophet
```

```python
m = Prophet()
```

```python
model = m.fit(df_prophet)
model
```

```python
# Perfomance evaluation
from prophet.diagnostics import cross_validation
```

```python
df_cv = cross_validation(model, horizon='30 days', period='15 days', initial='90 days')
```

```python
# Here is how it works. The initial model will be trained on the first 90 (because initial is set to 90 daya) days of data. 
# It will forecast the next 30 days of data (because horizon is set to 30). 
# The model will then train on the initial period + the period (90 + 15 days in this case) and forecast the next 30 days. 
# It will continued like this, adding another 15 days to the training data and then forecasting for the next 30 until there is no longer enough data to do this.

# In summary, period is how much data to add to the training data set in every iteration of cross-validation, and horizon is how far out it will forecast.
```

```python
df_cv
```

```python
from prophet.diagnostics import performance_metrics
```

```python
df_performance = performance_metrics(df_cv)
```
