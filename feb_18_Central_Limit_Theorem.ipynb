# Central limit theorem


The Central Limit Theorem (CLT) is a statistical theory that indicate
that the mean and standard deviation derived from a sample, 
will accurately approximate the mean and standard deviation of the population.
The minimum number of members in the sample is 30 or more than 30.
is 30

# law of large number

law of large number :
- sample size should be equal to or more than 30.
- Higher the sample size , lower the error .
- If the sample size is large enough more than 30 , the sample 
distribution taken from any population distribution regradless of 
its shape will approximately a normal distribution.

law of large number.

The sample size increases the mean of the sample is close to the mean of the population.

Create sample of size 200  plot histogram

```python
sampled_mean200=[]
for i in range(len(cardio_chol_df)):
    sampled_mean200.append(cardio_chol_df.sample(200).mean())

print(len(sampled_mean200))
sns.histplot(sampled_mean200,kde=True)
```

Create sample of size 100  plot histogram

```python
sampled_mean100=[]
for i in range(len(cardio_chol_df)):
    sampled_mean100.append(cardio_chol_df.sample(n=100).mean())
print(len(sampled_mean100))
sns.histplot(sampled_mean100,kde=True)
```

Create sample of size 30  plot histogram

```python
sampled_mean30=[]
for i in range(len(cardio_chol_df)):
    sampled_mean30.append(cardio_chol_df.sample(n=30).mean())
print(len(sampled_mean30))
sns.histplot(sampled_mean30,kde=True)
```

```python
# finding the error 
error=cardio_chol_df.mean()-pd.Series(sampled_mean200).mean()
error
```

```python
# finding the error 
error=cardio_chol_df.mean()-pd.Series(sampled_mean100).mean()
error
```

```python
# finding the error 
error=cardio_chol_df.mean()-pd.Series(sampled_mean30).mean()
error
```

If we see the above error closley as the sample size is decreasing error is increasing. In other words smaller the sample size larger the error
